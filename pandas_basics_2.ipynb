### Q22. ⭐ Interview Classic
When would you use `at` over `loc`? Write a comment, then use both to access the `salary` of the employee at index 4 and compare.

```python
# at is used to access or update a single value using row and column labels.
# It is generally faster than loc for retrieving one scalar value.
# loc is more flexible and can access one or multiple rows/columns.

# Using at
print(df.at[4, 'salary'])

# Using loc
print(df.loc[4, 'salary'])
```

### Q29. ⭐ Interview Classic
What happens when you use `loc` with a slice on a default integer index vs a custom index? Demonstrate by:
- Slicing `loc[1:3]` on the default index
- Setting `employee_id` as index and slicing `loc[103:107]`
- Write a comment explaining why the results differ.

```python
# Explanation: 
# loc always slices by LABELS, not positions.
# With the default integer index (0, 1, 2, ...), the labels happen to look
# like positions, so loc[1:3] returns rows with labels 1, 2, and 3.
# After setting employee_id as the index, the labels become employee IDs,
# so loc[103:107] returns all rows whose employee_id labels are between
# 103 and 107 (inclusive).
```
