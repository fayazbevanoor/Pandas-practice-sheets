### Real interview questions — think carefully before coding

### Q21. ⭐ Interview Classic
What is the difference between `dropna()` and `fillna()`? Write a comment, then show a case where dropping is better vs filling is better using this dataset.

```python
# Difference:
# dropna() removes rows or columns containing missing (NaN) values.
# fillna() replaces missing (NaN) values with a specified value, keeping the data.

# Example where dropping is better:
# Remove employees whose salary is missing.
df.dropna(subset=['salary'])

# Example where filling is better:
# Replace missing salaries with 50000 instead of removing those employees.
df['salary'].fillna(50000)
```

### Q22. ⭐ Interview Classic
Fill missing `salary` values with the **mean salary** of the entire column (not a hardcoded number).

```python
# df.describe().loc['mean','salary']
df['salary'].fillna(df.describe().loc['mean','salary'])


df['salary'].fillna(df['salary'].mean())
```

### Q23. ⭐ Interview Classic
Filter employees whose `name` contains the letter `a` (case-insensitive). Do not loop — use a pandas string method.

```python
df[df['name'].str.contains('a', case=False, na=False)]
```

### Q24. ⭐ Interview Classic
`dropna()` has a `thresh` parameter. Use it to keep only rows that have at least 11 non-null values.

### Q25. ⭐ Interview Classic
What does `duplicated(keep='last')` do differently from `duplicated(keep='first')`? Demonstrate with an example.

```python
# keep='first' (default):
# Marks duplicate rows as True except for the first occurrence.

# keep='last':
# Marks duplicate rows as True except for the last occurrence.

# Example
df_dup = pd.concat([df, df.iloc[[0]]], ignore_index=True)

print(df_dup.duplicated(keep='first'))
print(df_dup.duplicated(keep='last'))
```

### Q26. Find employees who have BOTH a missing salary AND less than 2 years of experience.

### Q29. ⭐ Interview Classic
Without hardcoding department names, filter all employees who belong to departments that have more than 4 employees. (Think: you need to count per department first.)

```python
dept_counts = df['department'].value_counts()
df[df['department'].isin(dept_counts[dept_counts > 4].index)]
```

### Q30. ⭐ Interview Classic
A common interview scenario: the dataset has missing salaries. Fill them using the **median salary of the employee's own department** (not the overall median). Each department should get its own median.

```python
df['salary'] = df['salary'].fillna(
    df.groupby('department')['salary'].transform('median')
)
```
