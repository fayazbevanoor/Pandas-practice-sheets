## 🔴 HARD (Q21–Q30)
### Real interview questions — think carefully before coding

### Q21. ⭐ Interview Classic
What is the difference between `sort_values()` and `nlargest()`? When would you prefer one over the other? Write a comment then demonstrate both to get the top 5 salaries.

```python
# sort_values():
# Sorts the entire DataFrame.

# nlargest():
# Efficiently returns only the top N rows based on a column without sorting the entire DataFrame.

# Using sort_values()
df.sort_values('salary', ascending=False).head(5)

# Using nlargest()
df.nlargest(5, 'salary')
```

### Q22. ⭐ Interview Classic
Sort the DataFrame by `salary` descending using `sort_values()` with `inplace=True`. What does `inplace=True` mean? Write a comment explaining it.

```python
# inplace=True means:
# It modifies the original DataFrame directly and returns None.

df.sort_values('salary', ascending=False, inplace=True)
```

### Q23. ⭐ Interview Classic
Find the highest paid employee in each department. Do not use `groupby` — only use `sort_values()` and `drop_duplicates()`.

```python
df.sort_values('salary', ascending=False).drop_duplicates(subset='department')
```

### Q24. ⭐ Interview Classic
Rank employees by salary within their department — the highest salary in each department gets rank 1. Use `sort_values()` with multiple columns.

```python
# df.sort_values(by=['department','salary'],ascending =[False,False])
df['salary_rank']= df.groupby('department')['salary'].rank(method='dense',ascending=False)

df
```

### Q25. Sort by `salary` descending and add a new column called `salary_rank` that numbers each row 1, 2, 3... after sorting.

```python
df=df.sort_values(by='salary',ascending=False).reset_index(drop=True)
df['salary_rank']=df.index+1
df
```

### Q26. ⭐ Interview Classic
Get the top 3 employees by `projects_completed` from EACH department separately. (Think: filter each department, nlargest, combine results.)

```python
top3 = (
    df.groupby('department', group_keys=False)
      .apply(lambda x: x.nlargest(3, 'projects_completed'))
)
top3
```

### Q27. ⭐ Interview Classic
Sort the DataFrame by `salary` but treat NaN salaries as if they were `0` during sorting — without actually filling them in the original data.

```python
df.sort_values('salary', key=lambda x: x.fillna(0), ascending=False)
```

### Q28. After sorting by `salary` descending, check if the index is still in its original order. Then reset it and verify.

```python
df_sorted = df.sort_values('salary', ascending=False)

print(df_sorted.index)

df_sorted = df_sorted.reset_index(drop=True)

print(df_sorted.index)
```

### Q29. ⭐ Interview Classic
Find the second highest salary in the dataset without using `nlargest()` directly — use `sort_values()` and iloc.

```python
df.sort_values(by='salary',ascending=False).iloc[1]

# df.nlargest(2,'salary').iloc[1]
```

### Q30. ⭐ Interview Classic
Sort employees by `experience_years` descending and `salary` descending. Then assign a `rank` column (1 being the most experienced and highest paid). Display the top 10.

```python
df= df.sort_values(by=['experience_years','salary'],ascending=[False,False]).reset_index(drop=True)
df['rank']=df.index+1
df.head(10)
```
