## 🟡 MEDIUM (Q11–Q20)

### Q11. Group by both `department` and `gender` and find the mean salary for each combination.

```python
df.groupby(['department','gender'])['salary'].mean()
```

### Q12. Group by `department` and find both the mean and max salary using `agg()`.

```python
df.groupby('department')['salary'].agg(['mean','max'])
```

### Q13. Group by `department` and calculate the mean `salary`, mean `rating`, and total `projects_completed` all at once using `agg()`.

```python
df.groupby('department').agg({
    'salary':'mean',
    'rating': 'mean',
    'projects_completed':'sum'
})
```

### Q14. Use `groupby` and `transform()` to add a new column `dept_avg_salary` that shows the average salary of each employee's department (same row count as original df).

```python
df['dept_avg_salary'] = df.groupby('department')['salary'].transform('mean')
df
```

### Q15. Group by `department` and use `agg()` with named aggregations to produce columns called `avg_salary` and `total_projects`.

```python
df.groupby('department').agg(
    avg_salary=('salary','mean'),
    total_project=('projects_completed','sum')
)
```

### Q16. Find the number of managers vs non-managers in each department using `groupby` on two columns.

```python
df.groupby(['department', 'is_manager']).size()
```

### Q17. Group by `department` and apply a custom function using `apply()` that returns the difference between the max and min salary in each group.

```python
df.groupby('department')['salary'].apply(lambda x: x.max() - x.min())
```

### Q18. Group by `city` and find how many unique departments exist in each city.

```python
df.groupby('city')['department'].nunique()
```

### Q19. Group by `department` and sort the result by average salary descending.

```python
df.groupby('department')['salary'].mean().sort_values(ascending=False)
```

### Q20. Group by `joining_year` and count how many employees joined each year. Sort by year ascending.

```python
df.groupby('joining_year')['employee_id'].count().sort_index(ascending=True)
```

## 🔴 HARD (Q21–Q30)
### Real interview questions — think carefully before coding

### Q21. ⭐ Interview Classic
What is the difference between `groupby` + `agg()` and `groupby` + `transform()`? Write a comment explaining the key difference in output shape, then demonstrate both.

```python
# agg():
# Returns one row per group (reduced output).

df.groupby('department')['salary'].agg('mean')

# transform():
# Returns the same number of rows as the original DataFrame.

df.groupby('department')['salary'].transform('mean')
```

### Q22. ⭐ Interview Classic
Add a column `salary_vs_dept_avg` that shows the difference between each employee's salary and their department's average salary. Use `transform()`.

```python
df['avg_dept_salary'] = df.groupby('department')['salary'].transform('mean')
df['diff'] = df['salary'] - df['avg_dept_salary']

df
```

### Q23. ⭐ Interview Classic
Find the top earner in each department. Use `groupby` and `apply()` only.

```python
df.groupby('department').apply(lambda x: x.loc[x['salary'].idxmax()])
```

### Q24. ⭐ Interview Classic
Group by `department` and compute a custom aggregation using `agg()` with a lambda that returns the range (max - min) of `experience_years` per department.

```python
df.groupby('department').agg(
    experience_range=('experience_years',
                      lambda x: x.max() - x.min())
)
```

### Q25. ⭐ Interview Classic
Find departments where the average salary is above the overall company average salary. Print only those department names.

```python
company_avg = df['salary'].mean()

df.groupby('department')['salary'].mean().loc[lambda x: x > company_avg].index
```

### Q26. ⭐ Interview Classic
Add a column `salary_rank_in_dept` that ranks each employee by salary within their department (rank 1 = highest). Use `groupby` and `rank()`.

```python
df['salary_rank_in_dept'] = df.groupby('department')['salary'].rank(method='dense',ascending=False)
df
```

### Q27. ⭐ Interview Classic
Group by `department` and use `agg()` to find — for each department — the mean salary, the count of employees, and the max rating, all in one call with custom column names.

```python
df.groupby('department').agg(
    avg_salary =('salary','mean'),
    employee_count = ('employee_id','sum'),
    max_rating= ('rating','max')
)
```

### Q28. ⭐ Interview Classic
For each department, find the employee with the most `projects_completed`. Return a DataFrame with department and that employee's name and project count.

```python
df.groupby('department').apply(
    lambda x: x.loc[x['projects_completed'].idxmax(),
                    ['name', 'projects_completed']]
)
```

### Q29. ⭐ Interview Classic
Use `groupby` and `filter()` to keep only the rows belonging to departments that have more than 4 employees.

```python
df.groupby('department').filter(lambda x: len(x) > 4)
```

### Q30. ⭐ Interview Classic
Group by `department` and `gender`. For each group find the mean salary. Then unstack the result so that gender becomes columns. This is a very common interview output format.

```python
df.groupby(['department', 'gender'])['salary'].mean().unstack()
```
